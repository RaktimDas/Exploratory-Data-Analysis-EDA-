```python
covid_df=covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

```python
covid_df.head()
```

```python
#Active Cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(data=covid_df,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc='max')
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100 / statewise['Confirmed']
```

```python
statewise['Death Rate'] = statewise['Deaths']*100 / statewise['Confirmed']
```

```python
statewise.head()
```

```python
statewise = statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='magma')
```

```python
"""Top 10 active covid cases in india """
top_10_cases = covid_df.groupby(['State/UnionTerritory']).max()[['Active_Cases','Date']].sort_values('Active_Cases',ascending=False).reset_index()
fig = plt.figure(figsize=(16, 9))
plt.title('Top 10 States with most active covid cases in India', size=20)
ax = sns.barplot(data=top_10_cases[:10], y='Active_Cases',x='State/UnionTerritory', palette='viridis', linewidth=3,edgecolor='black')
plt.xlabel('States', fontsize=15)
plt.ylabel('Total Active Cases', fontsize=15)
plt.show()
```

```python
"""Top 10 states with highest deaths"""
top_10_deaths = covid_df.groupby(['State/UnionTerritory']).max()[['Deaths','Date']].sort_values(by='Deaths',ascending=False).reset_index()
fig = plt.figure(figsize=(16, 9))
plt.title('Top 10 States with most  death cases in India', size=20)
ax = sns.barplot(data=top_10_deaths[:12], y='Deaths',x='State/UnionTerritory', palette='Set1', linewidth=3, edgecolor='black')
plt.xlabel('States', fontsize=15)
plt.ylabel('Total Death Cases',fontsize=15)
plt.show()
```

```python
"""Growth Trend"""
plt.figure(figsize=(14,8))
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka',
                  'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x='Date', y='Active_Cases', hue='State/UnionTerritory')
ax.set_title('Top 5 Affected States in India',size=16)
plt.show()
```

```python
"""Remove rows where state = India """
vaccine = covid_vaccine_df[covid_vaccine_df['State']!='India']
```

```python
vaccine.head()
```

```python
vaccine=vaccine.rename({'Total Individuals Vaccinated':'Total'},axis=1)
```

```python
vaccine.head()
```

```python
"""Most Vaccinated State"""
max_vac = vaccine.groupby('State')[['Total']].sum().sort_values('Total',ascending=False)[:5]
plt.figure(figsize=(12,6),dpi=100)
plt.title('Top 5 Vaccinated States in India',size=20)
sns.barplot(data=max_vac,y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
